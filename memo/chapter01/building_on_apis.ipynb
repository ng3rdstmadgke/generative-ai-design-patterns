# Anthropic Python SDK

```python
from anthropic.lib.bedrock import AnthropicBedrockMantle

client = AnthropicBedrockMantle(aws_region="us-east-1")

MODEL_ID="anthropic.claude-opus-4-8"
MODEL_ID = "anthropic.claude-haiku-4-5"
message = client.messages.create(
    # サポートされているモデル: https://platform.claude.com/docs/ja/build-with-claude/claude-in-amazon-bedrock#supported-models
    model=MODEL_ID,
    system="あなたはPythonプログラミングエキスパートです",
    max_tokens=1024,
    temperature=0.3,
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": "intの配列からmedianを取得するコードを書いて"
                }
            ]
        }
    ],
)

print(message.content[0].text)
```

# PydanticAI

```python
from pydantic_ai import Agent

agent = Agent(
    "bedrock:anthropic.claude-opus-4-8",
    system_prompt="あなたはPythonプログラミングエキスパートです",
)

result = agent.run_sync("intの配列からmedianを取得するコードを書いて")
print(result.data)
```
